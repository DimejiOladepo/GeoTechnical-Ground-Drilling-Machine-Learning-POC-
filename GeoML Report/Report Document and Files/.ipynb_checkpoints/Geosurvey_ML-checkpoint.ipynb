<h2> PHASE 1 - STRATEGY <h2>

---

<h3> PROBLEM DEFINITION <h3>

Geotechnical Ground Drilling is performed in many locations using Cone Penetration Test(CPD) where soil samples are taken at various depths(usually every 2.5cm) and are taken to a laboratory for testing. After laboratory tests, a laboratory report is drawn up where the soil type for a range of depths is determined and saved in a worksheet and each soil type encountered in that drill site is stored in a spreadsheet in the worksheet.

This led to an amassed data lake consisting of drill data files(.snd) and laboratory reports of the obtained soil samples(.xlsx/.xlsm).

                              drill data(.snd)                         workbook(.xlsx/.xlsm)
                              

<blockquote>  <i> The drill data consists of readings such as the longitudinal and latitudinal coordinates of the drill site, the height above sea level of the drill site, unique identifier, date drilling took place. The table consists of drill depth, drill pressure, drill torque and drill flushing pressure. 
    
The laboratory worksheet consists of several differently named spreadsheets with each containing the soil type at a range of depth. 
</i> </b> </blockquote>



<b> The question then became how could this stored data be mined in order to understand how soil type relates with other data and how this other data could "tell" what type of soil a sample is without having to do an expensive laboratory test.</b> 

___

<h3> PROBLEM SOLUTION <h3>

A Machine learning approach was considered to solve this problem where data would be fed into a suitable classification machine learning algorithm(Lightgbm) which will learn the relationship between the dependent variable (soil type) and other variables (drill depth, pressure etc.) and be able to predict the soil type of new unseen data given other variables accurately and swiftly. 
<ul>
<li> The first step was to create a dataset from the wealth of drill data and laboratory report files by extracting important data points in the drill data file and merging it with equally important data points in its sibling laboratory report. </li>

<li> The next step was to clean the new dataset by removing empty and incomplete rows and performing feature engineering through exploratory data analysis by balancing the dataset classes. </li> 

<li> The final step was then passing this clean and feature engineered dataset into an appropriately tuned machine learning algorithm where the algorithm would be able to predict soil type given other indicators(variables). </li>
</ul>

___

<h3> SCOPE DEFINITION <h3>

The program developed for this project will be able to access the directory where the data files are stored and be able to match each drill data file with its sibling laboratory report file in the same sub-directory and extract the important data points from each file, merge them together and repeat for all files in all sub-directories. Also, only spreadsheets named with integer numbers are considered in each workbook.

It will also clean up the output dataset and perform feature engineering techniques on it. After this, it will split the cleaned set into a training set and test/validation set and finally pass the train set through the Lightgbm ML algorithm to be able to predict soil type on the test/validation set and validate these predictions. It will also detail model metrics and results.

<h3> FEASABILITY DEFINITION <h3> 

This project would help save precious resources and time in determining the soil type at various locations and depths. This would enable faster decisions in managing soil and land in those locations.

```python
from IPython.display import display
from IPython.display import HTML
import IPython.core.display as di # Example: di.display_html('<h3>%s:</h3>' % str, raw=True)

# This line will hide code by default when the notebook is exported as HTML
di.display_html('<script>jQuery(function() {if (jQuery("body.notebook_app").length == 0) { jQuery(".input_area").toggle(); jQuery(".prompt").toggle();}});</script>', raw=True)

# This line will add a button to toggle visibility of code blocks, for use with the HTML export version
di.display_html('''<button onclick="jQuery('.input_area').toggle(); jQuery('.prompt').toggle();">Toggle code</button>''', raw=True)

import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn import metrics
from lightgbm import LGBMClassifier
import shap

dataset = pd.read_csv(r"data_extract.csv", index_col = 0)
dataset = dataset.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

y = dataset["Soil type"]
X = dataset.drop("Soil type", axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 50) 

gbm = LGBMClassifier(boosting_type = 'gbdt', colsample_bytree = 0.8, learning_rate = 0.1, 
                     n_estimators = 400, num_leaves = 20, objective = 'multiclass', 
                     random_state = 42, reg_alpha = 0.5, reg_lambda = 0.5, subsample = 0.75)

gbm.fit(X_train, y_train)
Y_sum = gbm.predict(X_test)

gridParams = {
    'learning_rate': [0.1, 0.05],
    'n_estimators': [40, 200, 400],
    'num_leaves': [20, 40],
    'boosting_type' : ['gbdt'],
    'objective' : ['multiclass'],
    'random_state' : [42], 
    'colsample_bytree' : [0.8, 1],
    'subsample' : [0.75,1],
    'reg_alpha' : [1,0.5],
    'reg_lambda' : [1,0.5],
    }

model = LGBMClassifier()
model.fit(X_train, y_train)

grid = GridSearchCV(model, gridParams,
                    verbose=0,
                    cv=4,
                    n_jobs=2)

grid.fit(X_train, y_train);
```

<h3> MODEL TRAINING <h3>

The Machine learning Algorithm chosen for model development was Lightgbm which is a gradient boosted tree which combines weak "learners" into a single strong learner in an iterative fashion. It has the advantage of being one of the fastest gradient boosted tree methods as well as being able to balance the soil type classes which was heavily skewed. 


```python
# This line will hide code by default when the notebook is exported as HTML
di.display_html('<script>jQuery(function() {if (jQuery("body.notebook_app").length == 0) { jQuery(".input_area").toggle(); jQuery(".prompt").toggle();}});</script>', raw=True)

# This line will add a button to toggle visibility of code blocks, for use with the HTML export version
di.display_html('''<button onclick="jQuery('.input_area').toggle(); jQuery('.prompt').toggle();">Toggle code</button>''', raw=True)

import lightgbm as lgb
_ = lgb.plot_tree(gbm, figsize = (50,50))
```

<blockquote> Above is a plot tree describing the flow of how the decision tree understands the relationship between the data features.</blockquote>


The algorithm works as follows:
Input: training set $\{(x_i,y_i)\}_{i=1}^n $ a differentiable loss function $\mathbf{L}(y_i,F(x))$, number of iterations M.

Algorithm:
<ol><i>
    <li>Initialize model with a constant value:</li>
$$F_0(x) = argmin_{\gamma}\sum_{i=1}^n L(y_i,\gamma)$$
    <li>For m = 1 to M:</li>
    <ol> 
        <li>Compute so-called pseudo-residuals:</li>
$$r_{im} = \left[ \frac{\partial L(y_i,F(x_i))}{\partial F(x_i)} \right]_{F(x)=F_{m-1(x)}}$$ for  i = 1,...,n 
<li>Fit a base learner (or weak learner, e.g. tree) $h_m(x)$  to pseudo-residuals, i.e. train it using the training set $\{(x_i,r_{im})\}_{i=1}^n$</li>
<li>Compute multiplier $\gamma_m$ by solving the following one-dimensional optimization problem:</li>
$$\gamma_m = argmin_\gamma \sum_{i=1}^n L(y_i,F_{m-1}(x_1)+\gamma h_m(x_i))$$
        <li>Update the model:</li>
$$F_m(x) = F_{m-1}(x)+\gamma_m h_m(x_i)).$$</ol>
    <li>Output $F_M(x)$.</li>
    </i></ol>
    
__[Read more in Wikipedia](https://en.wikipedia.org/wiki/Gradient_boosting)__
    
The model was trained on the training set using default parameters to predict the validation set. Here, the dependent column was dropped(i.e soil type) and fed into the model for prediction. The result is then checked against the the actual soil type column which was dropped.

<h3> IMPROVING MODEL PERFORMANCE BY HYPERPARAMETER TUNING <h3>

The developed model was then tuned to optimize for performance. This was done using GridSearchCV to search for the best parameters that could lead to an improvement in accuracy as well as other metrics. 

Here, a list of some parameter and hyper-parameter values were fed into the search algorithm were each was tested iteratively to obtain the best set of parameter and hyper-parameters for the model. 

Below is a visualization of the variation of mean test score with some parameters. 

```python
# This line will hide code by default when the notebook is exported as HTML
di.display_html('<script>jQuery(function() {if (jQuery("body.notebook_app").length == 0) { jQuery(".input_area").toggle(); jQuery(".prompt").toggle();}});</script>', raw=True)

# This line will add a button to toggle visibility of code blocks, for use with the HTML export version
di.display_html('''<button onclick="jQuery('.input_area').toggle(); jQuery('.prompt').toggle();">Toggle code</button>''', raw=True)

def plot_cv_results(cv_results, param_x, param_z, metric='mean_test_score'):
    """
    cv_results - cv_results_ attribute of a GridSearchCV instance (or similar)
    param_x - name of grid search parameter to plot on x axis
    param_z - name of grid search parameter to plot by line color
    """
    cv_results = pd.DataFrame(cv_results)
    col_x = 'param_' + param_x
    col_z = 'param_' + param_z
    fig, ax = plt.subplots(1, 1, figsize=(8, 5))
    sns.pointplot(x=col_x, y=metric, hue=col_z, data=cv_results, ci=99, n_boot=64, ax=ax)
    ax.set_title("CV Grid Search Results")
    ax.set_xlabel(param_x)
    ax.set_ylabel(metric)
    ax.legend(title=param_z)
    return fig

fig = plot_cv_results(grid.cv_results_, 'n_estimators', 'learning_rate'), 
plot_cv_results(grid.cv_results_, 'num_leaves', 'colsample_bytree')
```

<h3> MODEL EVALUATION AND TESTING <h3>

The model was then fitted and tested using the test set and this led to a precision score of 92%, recall of 88% and F-1 score of 89%. This means the model performed very well with a good ratio  of True Positives and True Negatives to False Positives and False Negatives. 


```python
# This line will hide code by default when the notebook is exported as HTML
di.display_html('<script>jQuery(function() {if (jQuery("body.notebook_app").length == 0) { jQuery(".input_area").toggle(); jQuery(".prompt").toggle();}});</script>', raw=True)

# This line will add a button to toggle visibility of code blocks, for use with the HTML export version
di.display_html('''<button onclick="jQuery('.input_area').toggle(); jQuery('.prompt').toggle();">Toggle code</button>''', raw=True)

#Model Accuracy
accuracy = metrics.accuracy_score(y_test, Y_sum)
print('Accuracy: %f' % accuracy)
#Model Precision
precision = metrics.precision_score(y_test, Y_sum, average = 'macro')
print('Precision: %f' % precision)
#Model Recall
recall = metrics.recall_score(y_test, Y_sum, average = 'macro')
print('Recall: %f' % recall)
#Model F1 score 
f1 = metrics.f1_score(y_test, Y_sum, average = 'macro')
print('F1 score: %f' % f1)

print("")
#Model Cross validation score 
print('Cross-Validation Score: %f' % np.mean(cross_val_score(gbm, X, y, cv=10)))
```

This means the model has the potential to do better with a lot more data points by using a larger dataset. This will mean a lot more SND and Workbook reports to be extracted as this would help improve performance and reduce overfitting.

<h3> STREAM LEARNING <h3>

The deployed ML model would then keep updating itself with new streams of data which would improve predictions and output. This would help ensure that the model is not stagnant (concept drift) and gives predictions based on only developmental data. 

The model would be exposed to a data source which will supply the new training data which will also contain the target variable. This process will be automated and occur at a selected number of times using limited computing and storage capabilities.
